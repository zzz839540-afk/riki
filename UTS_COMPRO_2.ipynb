```python
# Data Mentah dari sistem POS kafe Sou Space yang mengalami gangguan
raw_log = "ID:001#NAMA:kopi susu gula aren#QTY:3#PRICE:18000;ID:002#NAMA:COFFEE LATTE#QTY:1#PRICE:25000"

# 1. Memecah satu string menjadi dua transaksi
data = raw_log.split(";")

# Masing-masing transaksi dipecah menjadi 4 atribut terpisah
t1 = data[0].split("#")
t2 = data[1].split("#")

# 2. Membersihkan label teks dan spasi yang tidak diperlukan dari setiap atribut
# transaksi 1
nama1 = t1[1].replace("NAMA:","").strip()
qty1 = t1[2].replace("QTY:", "").strip()
price1 = t1[3].replace("PRICE:", "").strip()

# transaksi 2
nama2 = t2[1].replace("NAMA:","").strip()
qty2 = t2[2].replace("QTY:", "").strip()
price2 = t2[3].replace("PRICE:", "").strip()

# 3. Merapikan dan mengubah format tipe data siap pakai
# Merapikan nama produk agar huruf pertama tiap kata kapital
nama1 = nama1.title()
nama2 = nama2.title()

# Selanjutnya mengubah dengan cara konversi string ke integer supaya bisa dipakai untuk perhitungan matematis
qty1 = int(qty1)
qty2 = int(qty2)
price1 = int(price1)
price2 = int(price2)

# 4. Menghitung subtotal dan total pendapatan
sub1 = qty1 * price1
sub2 = qty2 * price2
total = sub1 + sub2

# 5. Hasil Laporan dengan format tabel yang rapi dan terstruktur
print("="*53)
print("LAPORAN HARIAN KEDAI SOU SPACE".center(50))
print("="*53)

print(f"{'MENU':<25}{'QTY':<8}{'HARGA':<10}{'SUBTOTAL':<10}")
print("-"*53)

print(f"{nama1:<25}{qty1:<8}{price1:<10}{sub1:<10}")
print(f"{nama2:<25}{qty2:<8}{price2:<10}{sub2:<10}")

print("-"*53)
print(f"{'TOTAL PENDAPATAN HARI INI:':<30}Rp {total}")
print("="*53)
```
